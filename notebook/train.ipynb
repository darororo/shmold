```python
from ultralytics import YOLO
import torch
import gc

try:
    model = YOLO("/home/eva/Documents/ITC/I5/ImageProcessing/mold_detection/shmold/runs/segment/train25/weights/best.pt") 
    
    metrics = model.val(data="../data.yaml", imgsz=640, conf=0.25)
    
    print(f"--- Benchmark Results ---")
    print(f"Precision (Quality): {metrics.results_dict['metrics/precision(B)']:.4f}")
    print(f"Recall (Completeness): {metrics.results_dict['metrics/recall(B)']:.4f}")
    print(f"mAP@50 (Overall Accuracy): {metrics.results_dict['metrics/mAP50(B)']:.4f}")
    print(f"mAP@50-95 (Strict Accuracy): {metrics.results_dict['metrics/mAP50-95(B)']:.4f}")
    
    speed = metrics.speed
    print(f"Inference Speed: {speed['inference']:.2f} ms per image")

finally:
    # Clean up GPU memory
    print("\n--- Cleaning up GPU memory ---")
    del model
    del metrics
    
    # Clear PyTorch cache
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
    
    # Force garbage collection
    gc.collect()
    
    print("GPU memory cleared successfully!")
```
