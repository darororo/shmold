```python
from ultralytics import YOLO

# 1. Load your model (YOLOv11, YOLOv12, or the new YOLO26)
model = YOLO("/home/eva/Documents/ITC/I5/ImageProcessing/mold_detection/shmold/runs/detect/train4/weights/best.pt") 

# 2. Run validation on your dataset (uses 'val' split by default)
# data="coco8.yaml" is a standard sample dataset for quick testing
metrics = model.val(data="../data.yaml", imgsz=640, conf=0.25)

# 3. Access and print the benchmarking metrics
print(f"--- Benchmark Results ---")
print(f"Precision (Quality): {metrics.results_dict['metrics/precision(B)']:.4f}")
print(f"Recall (Completeness): {metrics.results_dict['metrics/recall(B)']:.4f}")
print(f"mAP@50 (Overall Accuracy): {metrics.results_dict['metrics/mAP50(B)']:.4f}")
print(f"mAP@50-95 (Strict Accuracy): {metrics.results_dict['metrics/mAP50-95(B)']:.4f}")

# 4. Check inference speed (crucial for 2026 edge deployment)
speed = metrics.speed
print(f"Inference Speed: {speed['inference']:.2f} ms per image")
```
